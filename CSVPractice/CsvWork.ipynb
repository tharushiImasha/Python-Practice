WRITE

```python
import csv

data = [
    {"Name":"John", "Age":50, "City":"New York"},
    {"Name":"Jane", "Age":30, "City":"Colombo"},
    {"Name":"Bob", "Age":25, "City":"Chicago"},
    {"Name":"Kate", "Age":32, "City":"Los Angeles"}
]

field_names = ["Name", "Age", "City"]

with open ('my_csv_file.csv', 'w', newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=field_names)

    csv_writer.writeheader()

    for row in data:
        csv_writer.writerow(row)
```

```python
import csv

with open ('employees.csv', 'r', newline='') as file:
    csv_reader = csv.reader (file)

    for row in csv_reader:
        print(row)
        print(row[3])

filterd_data = []

with open ('employees.csv', 'r', newline='') as file:
    csv_reader = csv.DictReader (file)

    for row in csv_reader:
        salary = int(row["Salary"])
        if salary >= 60000:
            print(row)
            filterd_data.append(row)

field_names = ["EmployeeID","Name","Departmanet","Salary"]

with open ('high_salary_employees.csv', 'w', newline='') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=field_names)

    csv_writer.writeheader()

    for row in filterd_data:
        csv_writer.writerow(row)   
```
