```python
import requests

response = requests.get('https://api.github.com/events')

if response.status_code == 200:
    print("Successfully fetched the page")
    
    if 'application/json' in response.headers['Content-Type']:
        data = response.json()
        print(data)
        print(type(data))
        
        for key in data[1].keys():
            print(key)
    
else:
    print("Some error happend while fetching the page")
```

POST

```python
import requests

data = { 'user_name':'user', 'password': 'pass'}

url = 'https://httpbin.org/post'

response = requests.post(url, data)

if response.status_code == 200:
    print("Successfully access")
    print(response.text)
else:
    print("Login failed")
    print(response.status_code)
```

GET FILTERD DATA

```python
import requests

url = 'https://example.com'

params = {'q': 'pyhton', 'category':'books'}

response = requests.get(url, params=params)

if response.status_code == 200:
    print("Successfully fetched the page")
    print(response.url)
    print(response.text)
    
else:
    print("Some error happend while fetching the page")
```

BASIC AUTH

```python
import requests

url = 'https://httpbin.org/get'

user_name = 'user'
password = 'pass'

response = requests.get(url, auth=(user_name, password))

if response.status_code == 200:
    print("Successfully fetched the page")
    print(response.text)
    
else:
    print("Some error happend while fetching the page")
```
